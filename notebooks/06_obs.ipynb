# Show some of the obs/reanalysis datasets available

* Load CERES_EBAF, ERA5 and IMERG

```python
import intake
import pandas as pd

import easygems.healpix as egh
```

```python
# Filter out annoying warning.
import warnings
warnings.filterwarnings("ignore", message=".*The return type of `Dataset.dims` will be changed.*", category=FutureWarning)
```

```python
# Open catalog.
url = 'https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml'
cat = intake.open_catalog(url)['UK']
# Use online if not on JASMIN.
# cat = intake.open_catalog('https://digital-earths-global-hackathon.github.io/catalog/catalog.yaml')['online']
```

```python
ds_ceres = cat['CERES_EBAF'].to_dask()
```

```python
ds_era5 = cat['ERA5'](zoom=8).to_dask()
```

```python
ds_imerg = cat['IR_IMERG'](zoom=9).to_dask()
```

* what frequency are each of these datasets?

```python
ds_ceres.time
```

```python
ds_era5.time
```

```python
ds_imerg.time
```

* Given the different frequencies of each dataset, why would you have to be careful using this time in the `ds.sel(...)` call below?

```python
time = pd.Timestamp('2021-05-18 13:00')
```

```python
egh.healpix_show(ds_ceres.sel(time=time, method='nearest').toa_lw_all_mon)
```

```python
egh.healpix_show(ds_era5.sel(time=time, method='nearest').tcwv)
```

```python
egh.healpix_show(ds_imerg.sel(time=time, method='nearest').Tb)
```
