```python
import xarray as xr
import matplotlib.pyplot as plt
import warnings
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np

%xmode Plain
```

**N2560 CoMA9**\
LMI: 2020-09-29T00:00 (20.02, 129)\
Peak IR: 2020-09-28T06:00 (19.35, 135)\
RI onset: 2020-09-27T06:00 (16.55, 141.8)

##### 5km CoMA9

```python
diri = '/work/scratch-pw4/emg/LoSSETT/output/DYAMOND3/DYAMOND3/'
var = 'inter_scale_transfer_of_kinetic_energy'
date = '20200927T00'
model = 'glm.n2560_CoMA9_hier_v2'
fili = f'{model}_n2560_{var}_Lmin_00008_Lmax_00250_{date}_50S-50N_p0*'

ds = xr.open_mfdataset(diri+fili)
```

```python
CM5_TC = ds['Dl_u'].sel(latitude=16.55, method='nearest', length_scale = 125000, time = dt.datetime(2020,9,27,6))
CM5_TC_slice = CM5_TC.sel(longitude=slice(138,146))
```

```python
fig,ax = plt.subplots(1,1)

levs = np.linspace(-5e-2, 5e-2, 21)

cb = ax.contourf(CM5_TC_slice.longitude, CM5_TC_slice.pressure, CM5_TC_slice, cmap='RdBu_r', levels = levs, extend='both')
CB = fig.colorbar(cb)
CB.set_label(r'ISET (m$^2$ s$^{-3}$)')
ax.invert_yaxis()
ax.set_yscale('log')

ax.set_title('5km CoMA9 L = 250 km 24h RI onset')
ax.set_ylabel('Pressure (hPa)')
ax.set_xlabel(r'Longitude $^\circ$E')

pressure_levels = CM5_TC_slice.pressure.values
ax.set_yticks(pressure_levels)
ax.set_yticks([], minor=True)  # remove minor ticks
ax.yaxis.set_major_formatter(plt.ScalarFormatter())
ax.yaxis.set_minor_formatter(plt.NullFormatter())  # suppress minor labels
ax.set_yticklabels([str(p) for p in pressure_levels])
ax.set_ylim(851, 100)

fig.savefig('/home/users/emg97/emgPlots/hk26_5km_CoMA9_L250km_Dlu_slice_TC_RIonset.png', bbox_inches='tight')
```
